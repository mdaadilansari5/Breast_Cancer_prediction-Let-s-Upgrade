```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import  StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the csv file
df=pd.read_csv("https://raw.githubusercontent.com/ingledarshan/AIML-B2/main/data.csv")
df.head()
```

```python
#Using dropna to drop out the Unnamed: 32 column
df.drop("Unnamed: 32", axis=1, inplace=True)
```

# Data exploration

```python
df.describe()
```

```python
# Splitting the dataset into independent variable and target variable
X=df.drop('diagnosis',axis=1)
print(X)
y=df['diagnosis']
y
```

```python
X.head()
```

```python
y.head(5)
```

```python
#Using LabelEncoder to convert text data into numeral values
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
y
```

```python
# Scaling down the X dataset (Feature Scaling) using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
X=ms.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test.shape
```

# Machine Learning Models and Comparitive Analysis

### Logistic Regression

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
# Checking the accuracy of y_pred with respect to y_test
from sklearn.metrics import accuracy_score
lr_acc=accuracy_score(y_pred,y_test)
print(lr_acc)
```

```python
results = pd.DataFrame()
tempResults = pd.DataFrame({'Algorithm':['Logistic Regression Method'], 'Accuracy':[lr_acc]})
results = pd.concat( [results, tempResults] )
results
```

### Decision Tree Classifier

```python
dtf=DecisionTreeClassifier()
dtf.fit(X_train,y_train)
```

```python
y_pred=dtf.predict(X_test)
```

```python
dtf_acc=accuracy_score(y_pred,y_test)
print(dtf_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Decision Tree Classifier'], 'Accuracy':[dtf_acc]})
results = pd.concat( [results, tempResults] )
results
```

### Random Forest Classifier

```python
rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
y_pred=rfc.predict(X_test)
```

```python
rfc_acc=accuracy_score(y_pred,y_test)
print(rfc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Random Forest Classifier'], 'Accuracy':[rfc_acc]})
results = pd.concat( [results, tempResults] )
results
```

### Support Vector Classifier

```python
svc=SVC()
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
svc_acc=accuracy_score(y_pred,y_test)
print(svc_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Support Vector Classifier'], 'Accuracy':[svc_acc]})
results = pd.concat( [results, tempResults] )
results
```

##### We can observe that Support Vector Classifier has the maximum accurary score of around 97%,i.e; it has predicted 97% accurate of the cancer type as apposed to test dataset
